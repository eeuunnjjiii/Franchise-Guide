## 4) 형변환

```python
# 콤마 제거 후 실수형 변환
def drop_comma_to_float(num):
    if type(num) == int:
      return num
    else:
      num = num.replace(',','')
      return float(num)

# '명' 제거 후 실수형 변환
def drop_str_to_float(num):
    if type(num) == int: ## 0은 그대로 유지
      return num
    else:
      num = num.replace('명','')
      num = num.replace(',','')
      return float(num)

# '\n' 제거 후 실수형 변환
def drop_enter_to_float(num):
    if type(num) == int: ## 0은 그대로 유지
      return num
    else:
      num = num.split('\n')[0]
      num = num.replace(',','')
      return float(num)

# 날짜를 datetime으로 변경하는 함수
def to_datetime(date):
    try:
        new_time = pd.to_datetime(date).date()
        return new_time
    except:
        return date
```

```python
for i in ['가맹점평균매출액','면적당평균매출액','합계']:
    brand[i] = brand[i].apply(drop_enter_to_float)

for i in ['가맹비','교육비','보증금','기타비용','면적당인테리어비용','기준면적당인테리어비용','총인테리어비용']:
    brand[i] = brand[i].apply(drop_comma_to_float)

brand['가맹사업개시일'] = brand['가맹사업개시일'].apply(to_datetime)
brand['가맹본부임직원수'] = brand['가맹본부임직원수'].apply(drop_str_to_float)
```

```python
# 가맹사업년수를 days로 변경하는 함수
def to_days_2020(date):
    try:
      days = (pd.to_datetime('2020-1-1').date()-date).days
      return days
    except:
      return 0

def to_days_2019(date):
    try:
      days = (pd.to_datetime('2019-1-1').date()-date).days
      return days
    except:
      return 0

def to_days_2018(date):
    days = (pd.to_datetime('2018-1-1').date()-date).days
    return days

def to_days_2017(date):
    days = (pd.to_datetime('2017-1-1').date()-date).days
    return days

def to_days_2016(date):
    days = (pd.to_datetime('2016-1-1').date()-date).days
    return days
```
