```python
import pandas as pd

# 데이터 가져오기
growth = pd.read_csv('/content/drive/MyDrive/franchise_data/growth.csv', index_col=0)
stability = pd.read_csv('/content/drive/MyDrive/franchise_data/stability.csv', index_col=0)
profit = pd.read_csv('/content/drive/MyDrive/franchise_data/profit.csv', index_col=0)
overview = pd.read_csv('/content/drive/MyDrive/franchise_data/overview.csv', index_col=0)
info = pd.read_csv('/content/drive/MyDrive/franchise_data/info.csv', index_col=0)
cost = pd.read_csv('/content/drive/MyDrive/franchise_data/cost.csv', index_col=0)

# 데이터 확인
growth.shape, stability.shape, profit.shape, overview.shape, info.shape, cost.shape
```

# 1. 가맹본부별 데이터 통합

```python
head_office = growth.merge(stability, on=['상호','자산','업종','기준년도'], how='left')
head_office = head_office.merge(profit, on=['상호','자본','매출액','영업이익','업종','기준년도'], how='left')
print(head_office.shape)
head_office.head(3)
```

## 1) 문자로 표시된 특성 삭제

```python
head_office = head_office.drop(['총자산증가율','매출액증가율','영업이익증가율'], axis=1)
head_office.head(3)
```

## 2) 컬럼명 변경

```python
# 컬럼명 확인
head_office.columns
```

```python
# 컬럼명 변경
head_office.columns = ['상호', '자산', '매출액', '영업이익', '업종', '기준년도', '자본', '부채', '부채비율',
                      '자기자본비율', '법위반횟수', '당기순이익', '영업이익률','매출액순이익률', '자기자본순이익률']
head_office.head(3)
```

## 3) 결측치 확인

```python
head_office.isna().sum()
```

```python
# 결측치 확인
head_office[head_office['자산'].isna()]
```

- 자산이 0인 경우 영업이익도 0인 경우가 많으므로 삭제
- 분모에 해당하는 매출액이 0인 경우 영업이익률, 매출액순이익률, 자기자본순이익률이 NaN으로 표시됨 => NaN 값 0으로 대체

```python
# 자산이 결측치인 데이터 삭제
head_office = head_office.dropna(subset=['자산'])
head_office.shape
```

```python
# 나머지 결측치 0으로 변환
head_office = head_office.fillna(0)
```

```python
# 값이 공백인경우 0으로 변환
def change(value):
  if value == '%':
    return 0
  else:
    return value
head_office['부채비율'] = head_office['부채비율'].apply(change)
head_office['자기자본비율'] = head_office['자기자본비율'].apply(change)
head_office.shape
```

## 4) 중복값 확인

```python
head_office.duplicated().sum()
```

```python
# 중복값 확인
head_office[head_office.duplicated()]
```

```python
# 중복값 제거
head_office = head_office.drop_duplicates(subset=['상호','자산','업종','기준년도'], keep='last')
head_office.shape
```

## 5) 형변환

```python
# 콤마 제거 후 실수형 변환
def drop_comma_to_float(num):
    if type(num) == int:
      return num
    else:
      num = num.replace(',','')
      return float(num)

# % 제거 후 실수형 변환
def drop_per_to_float(num):
    if type(num) == int: ## 0은 그대로 유지
      return num
    else:
      num = num.replace('%','')
      num = num.replace(',','')
      return float(num)
```

```python
for i in ['자산','매출액','영업이익','자본','부채','당기순이익','영업이익률','매출액순이익률','자기자본순이익률']:
    head_office[i] = head_office[i].apply(drop_comma_to_float)

for i in ['부채비율','자기자본비율']:
    head_office[i] = head_office[i].apply(drop_per_to_float)
```

```python
head_office = head_office[['상호', '업종', '기준년도', '자산', '자본', '부채', '매출액', '영업이익', '당기순이익', 
                           '부채비율', '자기자본비율','영업이익률', '매출액순이익률', '자기자본순이익률','법위반횟수']]
```

```python
# 정수형 변경
for i in head_office.columns[3:9]:
  head_office[i] = head_office[i].apply(int)
```

```python
# 데이터 확인
print(head_office.info())
head_office.head(3)
```

# 2. 브랜드별 데이터 통합

```python
brand = overview.merge(info, on=['브랜드','상호','업종','기준년도'], how='left')
brand = brand.merge(cost, on=['브랜드','상호','업종','기준년도'], how='left')
print(brand.shape)
brand.head(3)
```

## 1) 컬럼명 변경

```python
# 중복되는 가맹점수 컬럼 삭제
brand = brand.drop('가맹점수_y', axis=1)

# 브랜드명이 '평균'인 데이터 삭제
ls = brand[brand['브랜드']=='평균'].index.to_list()
brand = brand.drop(ls)
```

```python
# 컬럼명 확인
brand.columns
```

```python
# 컬럼명 변경
brand.columns = ['브랜드', '상호', '가맹사업개시일', '가맹사업년수', '가맹점수', '가맹본부임직원수', '업종', '기준년도',
                 '신규개점', '계약종료', '계약해지', '명의변경', '가맹점평균매출액','면적당평균매출액', '가맹비', '교육비', '보증금',
                 '기타비용', '합계', '면적당인테리어비용', '기준면적당인테리어비용', '총인테리어비용']
brand.head(3)
```

## 2) 결측치 확인

```python
# 결측치 확인
brand[brand['가맹비'].isna()]
```

- 결측치는 전부 0으로 대체

```python
# 결측치 0으로 변환
brand = brand.fillna(0)
```

```python
brand.duplicated(subset=['브랜드','상호','합계','업종','기준년도']).sum()
```

```python
# 중복값 확인
brand[brand.duplicated(subset=['브랜드','상호','합계','업종','기준년도'])]
```

```python
# 중복값 제거
brand = brand.drop_duplicates(subset=['브랜드','상호','합계','업종','기준년도'], keep='last')
brand.shape
```

## 4) 형변환

```python
# 콤마 제거 후 실수형 변환
def drop_comma_to_float(num):
    if type(num) == int:
      return num
    else:
      num = num.replace(',','')
      return float(num)

# '명' 제거 후 실수형 변환
def drop_str_to_float(num):
    if type(num) == int: ## 0은 그대로 유지
      return num
    else:
      num = num.replace('명','')
      num = num.replace(',','')
      return float(num)

# '\n' 제거 후 실수형 변환
def drop_enter_to_float(num):
    if type(num) == int: ## 0은 그대로 유지
      return num
    else:
      num = num.split('\n')[0]
      num = num.replace(',','')
      return float(num)

# 날짜를 datetime으로 변경하는 함수
def to_datetime(date):
    try:
        new_time = pd.to_datetime(date).date()
        return new_time
    except: ## '가맹사업 개시전'은 21년 1월 1일로 지정
        return pd.to_datetime('2021-1-1').date()
```

```python
for i in ['가맹점평균매출액','면적당평균매출액','합계']:
    brand[i] = brand[i].apply(drop_enter_to_float)

for i in ['가맹비','교육비','보증금','기타비용','면적당인테리어비용','기준면적당인테리어비용','총인테리어비용']:
    brand[i] = brand[i].apply(drop_comma_to_float)

brand['가맹사업개시일'] = brand['가맹사업개시일'].apply(to_datetime)
brand['가맹본부임직원수'] = brand['가맹본부임직원수'].apply(drop_str_to_float)
```

```python
# 가맹사업년수를 days로 변경하는 함수
def to_days(date):
    days = (pd.to_datetime('2021-1-1').date()-date).days
    return days

brand['가맹사업일수'] = brand['가맹사업개시일'].apply(to_days)

# 가맹사업년수 컬럼 삭제
brand = brand.drop('가맹사업년수', axis=1)
```

```python
brand = brand[['브랜드', '상호', '업종', '기준년도', '가맹사업개시일', '가맹사업일수', '가맹점수', '가맹본부임직원수',
               '신규개점','계약종료', '계약해지', '명의변경', '가맹점평균매출액', '면적당평균매출액', '가맹비', '교육비', '보증금',
               '기타비용', '합계', '면적당인테리어비용', '기준면적당인테리어비용', '총인테리어비용']]
```

```python
# 정수형 변경
for i in brand.columns[5:]:
  brand[i] = brand[i].apply(int)
```

```python
# 데이터 확인
print(brand.info())
brand.head(3)
```
