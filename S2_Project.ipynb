```python
#파일명 지정
summary=pd.read_csv('브랜드별개요.csv')
sales=pd.read_csv('가맹점현황정보.csv')
cost=pd.read_csv('가맹점창업비용.csv')
head_safety=pd.read_csv('가맹본부별자료_안정성.csv')
head_profit=pd.read_csv('가맹본부별자료_수익성.csv')
head_growth=pd.read_csv('가맹본부별자료_성장성.csv')
```

### 데이터 전처리

```python
#중복값 제거
summary=summary.drop_duplicates()
sales=sales.drop_duplicates()
cost=cost.drop_duplicates()
head_safety=head_safety.drop_duplicates()
head_profit=head_profit.drop_duplicates()
head_growth=head_growth.drop_duplicates()
```

#### summary 파일 전처리

```python
import warnings
warnings.filterwarnings(action='ignore')

#가맹사업 개시일, 가맹사업 년수 수정
summary=summary[summary['가맹사업 개시일']!='0']

#기준년도 2020년
s_2020=summary[summary['기준연도']==2020]
s_2020['year']='2020-12-31'
s_2020['year']=pd.to_datetime(s_2020['year'])
s_2020['가맹사업 개시일']=pd.to_datetime(s_2020['가맹사업 개시일'])
s_2020['days']=s_2020['year']-s_2020['가맹사업 개시일']

#기준년도 2019년
s_2019=summary[summary['기준연도']==2019]
s_2019['year']='2019-12-31'
s_2019['year']=pd.to_datetime(s_2019['year'])
s_2019['가맹사업 개시일']=pd.to_datetime(s_2019['가맹사업 개시일'])
s_2019['days']=s_2019['year']-s_2019['가맹사업 개시일']

#기준년도 2018년
s_2018=summary[summary['기준연도']==2018]
s_2018['year']='2018-12-31'
s_2018['year']=pd.to_datetime(s_2018['year'])
s_2018['가맹사업 개시일']=pd.to_datetime(s_2018['가맹사업 개시일'])
s_2018['days']=s_2018['year']-s_2018['가맹사업 개시일']

#기준년도 2017년
s_2017=summary[summary['기준연도']==2017]
s_2017['year']='2017-12-31'
s_2017['year']=pd.to_datetime(s_2017['year'])
s_2017['가맹사업 개시일']=pd.to_datetime(s_2017['가맹사업 개시일'])
s_2017['days']=s_2017['year']-s_2017['가맹사업 개시일']
```

```python
#다시 합치기
summary=pd.concat([s_2020,s_2019])
summary=pd.concat([summary,s_2018])
summary=pd.concat([summary,s_2017])
summary=summary.drop(['가맹사업 개시일','가맹사업 년수','year'],axis=1)
summary.columns=['업종','브랜드','상호','가맹점수','가맹본부임직원수','기준연도','가맹개시일수']
summary.head(3)
```

```python
#가맹개시일수 days 정수형 변경
days=summary['가맹개시일수'].astype('str').tolist()

for i in range(len(days)):
    days[i]=days[i].split('days')[0]

summary['가맹개시일수']=days
summary['가맹개시일수']=summary['가맹개시일수'].astype(int)

#가맹본부임직원수 정수형 변경
summary['가맹본부임직원수']=summary['가맹본부임직원수'].str.replace('명','')
print(summary.shape)
summary.head()
```

#### sales 파일 전처리

```python
#정수형 콤마 제거
sales['가맹점\n평균매출액']=sales['가맹점\n평균매출액'].str.replace(',','').astype(float)
sales['가맹점\n면적(3.3㎡)당\n평균매출액']=sales['가맹점\n면적(3.3㎡)당\n평균매출액'].str.replace(',','').astype(float)
```

```python
#평균매출액이 0이면 삭제
sales=sales[sales['가맹점\n평균매출액']>0]
print(sales.shape)
sales.head(3)
```

#### cost 파일 전처리

```python
#콤마 제거
cost['가입비\n(가맹비)']=cost['가입비\n(가맹비)'].str.replace(',','').astype(float)
cost['교육비']=cost['교육비'].str.replace(',','').astype(float)
cost['보증금']=cost['보증금'].str.replace(',','').astype(float)
cost['기타비용\n(인테리어\n비용포함)']=cost['기타비용\n(인테리어\n비용포함)'].str.replace(',','').astype(float)
cost['인테리어비용 면적당(3.3㎡)\n비용']=cost['인테리어비용 면적당(3.3㎡)\n비용'].str.replace(',','').astype(float)
cost['인테리어비용 기준면적\n(㎡)']=cost['인테리어비용 기준면적\n(㎡)'].str.replace(',','').astype(float)
cost['인테리어비용 총 비용\n']=cost['인테리어비용 총 비용\n'].str.replace(',','').astype(float)
```

```python
#합계 데이터 정수형으로 변경
total_list=cost['합계\n(창업비용\n지수)'].tolist()
try:
  for i in range(len(total_list)+1):
    total_list[i]=total_list[i].split('\n')[0].replace(',','')
except:
  pass
cost['합계\n(창업비용\n지수)']=total_list
cost['합계\n(창업비용\n지수)']=cost['합계\n(창업비용\n지수)'].str.replace(',','').astype(float)
```

```python
#잘못 입력된 데이터 삭제
cost=cost.drop([16560,15983,1,12504,0])
print(cost.shape)
cost.head()
```

#### head 파일 전처리

```python
#부채비율, 자기자본비율 % 제거
head_safety['부채비율\n(부채/자본)']=head_safety['부채비율\n(부채/자본)'].str.replace('%','')
head_safety['자기자본비율\n(자본/자산)']=head_safety['자기자본비율\n(자본/자산)'].str.replace('%','')

#자산,자본,부채 콤마 제거
head_safety['자산']=head_safety['자산'].str.replace(',','').astype('float')
head_safety['자본']=head_safety['자본'].str.replace(',','').astype('float')
head_safety['부채']=head_safety['부채'].str.replace(',','').astype('float')
head_growth['자산']=head_growth['자산'].str.replace(',','').astype('float')
head_growth['매출액']=head_growth['매출액'].str.replace(',','').astype('float')
head_growth['영업이익']=head_growth['영업이익'].str.replace(',','').astype('float')
head_profit['자본']=head_profit['자본'].str.replace(',','').astype('float')
head_profit['매출액']=head_profit['매출액'].str.replace(',','').astype('float')
head_profit['영업이익']=head_profit['영업이익'].str.replace(',','').astype('float')
head_profit['당기순이익']=head_profit['당기순이익'].str.replace(',','').astype('float')
```

```python
#매출액 컬럼 0이하 삭제
head_growth=head_growth[head_growth['매출액']>0]
head_profit=head_profit[head_profit['매출액']>0]
```

## 데이터 랭글링

#### 가맹본부별 자료_안정성,성장성,수익성 merge

```python
head_safety.shape, head_profit.shape, head_growth.shape
```

```python
head=pd.merge(head_safety,head_profit,on=['상호','기준연도','업종'],how='inner')
head=head.merge(head_growth,on=['상호','기준연도','업종'],how='inner')
#columns 정리
head=head.drop(['자본_y','자산_y', '매출액_y', '영업이익_y'],axis=1)
head.columns=['업종', '상호', '자산', '자본', '부채', '부채비율', '자기자본비율',
       '법위반횟수', '기준연도', '매출액', '영업이익', '당기순이익', '영업이익률', '매출액순이익률', '자기자본순이익률', '총자산증가율', '매출액증가율', '영업이익증가율']
print(head.shape)
head.head(3)
```

#### 가맹본부별 자료 & 브랜드별 개요 merge

```python
summary_head=pd.merge(summary,head,how='left',on=['상호','기준연도','업종'])#head&summary merge
print(summary_head.shape)
summary_head.head(3)
```

#### 가맹본부별 자료 & 브랜드별 개요 &  가맹점 창업비용 merge

```python
summary_head_cost=summary_head.merge(cost,how='left',on=['브랜드','상호','기준연도','업종'])
print(summary_head_cost.shape)
summary_head_cost.head(3)
```

#### 가맹본부별 자료 & 브랜드별 개요 &  가맹점 창업비용 & 연도별 가맹점현황정보 merge

```python
data=summary_head_cost.merge(sales,how='left',on=['브랜드','기준연도','업종'])
data=data.drop(['상호_y','가맹점수_y'],axis=1) #columns 정리
print(data.shape)
data.head()
```

```python
def eda(df):
    
    #정보공개서에서 쉽게 열람할 수 없는 특성 삭제
    df=df.drop(['부채비율','자기자본비율','영업이익률','매출액순이익률','자기자본순이익률',
             '총자산증가율','매출액증가율','영업이익증가율'],axis=1)

    #data leakage 특성 제거
    df=df.drop(['평균매출액','평균매출액_면적당','업종별평균매출액'],axis=1)

    #data cardinality 높은 특성 제거
    df =df.drop(['브랜드','상호'], axis=1)
    
    #중복되는 특성 제거
    df= df.drop(['인테리어비용_기준면적','인테리어비용_면적당','기준연도'],axis=1)
    
    #합계(가맹비+교육비+보증금+기타비용)에 해당되는 특성 제거
    df=df.drop(['가맹비','교육비','보증금','기타비용'],axis=1)
    
    return df

X_train=eda(X_train)
X_val=eda(X_val)
X_test=eda(X_test)
```

```python
def explain(row_number):
    positive_class = '1'
    positive_class_index = 1

    # row 값을 변환합니다
    row = X_test.iloc[[row_number]]
    row_processed = encoder.transform(row)

    # 예측하고 예측확률을 얻습니다 
    pred = model.predict(row_processed)[0]
    pred_proba = model.predict_proba(row_processed)[0, positive_class_index]
    pred_proba *= 100
    #if pred != positive_class:
    #    pred_proba = 100 - pred_proba
    if pred ==0:
        pred='"평균미만"'
    else:
        pred='"평균이상"'

    # 예측결과와 확률값을 얻습니다
    print(f'이 브랜드에 대한 예측결과는 {pred} 으로, 확률은 {pred_proba:.0f}% 입니다.')
    
    # SHAP를 추가합니다
    shap_values = explainer.shap_values(row_processed)

    # Fully Paid에 대한 top 3 pros, cons를 얻습니다
    feature_names = row.columns
    feature_values = row.values[0]
    shaps = pd.Series(shap_values[0], zip(feature_names, feature_values))
    pros = shaps.sort_values(ascending=False)[:3].index
    cons = shaps.sort_values(ascending=True)[:3].index

    # 예측에 가장 영향을 준 top3
    print('\n')
    print('Positive 영향을 가장 많이 주는 3가지 요인 입니다:')
    
    evidence = pros if pred == positive_class else cons
    for i, info in enumerate(evidence, start=1):
        feature_name, feature_value = info
        print(f'{i}. {feature_name} : {feature_value}')

    # 예측에 가장 반대적인 영향을 준 요인 top1
    print('\n')
    print('Negative 영향을 가장 많이 주는 3가지 요인 입니다:')
    
    evidence = cons if pred == positive_class else pros
    for i, info in enumerate(evidence, start=1):
        feature_name, feature_value = info
        print(f'{i}. {feature_name} : {feature_value}')

    # SHAP
    shap.initjs()
    return shap.force_plot(
        base_value=explainer.expected_value, 
        shap_values=shap_values, 
        features=row, 
        link='logit'
    )
```

```python
#기준모델 생성
from sklearn.metrics import mean_absolute_error,r2_score
y_pred=[y_train.mean()]*len(y_train)
mae=mean_absolute_error(y_train,y_pred)
print('기준모델 MAE:',mae)
```

```python
def eda3(df):
    
    #정보공개서에서 쉽게 열람할 수 없는 특성 삭제
    df=df.drop(['부채비율','자기자본비율','영업이익률','매출액순이익률','자기자본순이익률',
             '총자산증가율','매출액증가율','영업이익증가율'],axis=1)

    #data leakage 특성 제거
    df=df.drop(['평균매출액_면적당','업종별평균매출액','평가'],axis=1)

    #data cardinality 높은 특성 제거
    df =df.drop(['브랜드','상호'], axis=1)
    
    #중복되는 특성 제거
    df= df.drop(['인테리어비용_기준면적','인테리어비용_면적당','기준연도'],axis=1)
    
    #합계(가맹비+교육비+보증금+기타비용)에 해당되는 특성 제거
    df=df.drop(['가맹비','교육비','보증금','기타비용'],axis=1)
    return df

X_train=eda3(X_train)
X_val=eda3(X_val)
X_test=eda3(X_test)
```

```python
#early stoping

encoder = TargetEncoder()
X_train_encoded = encoder.fit_transform(X_train,y_train) # 학습데이터
X_val_encoded = encoder.transform(X_val) # 검증데이터

model = XGBRegressor(
    n_estimators=1000, 
    max_depth=7,
    learning_rate=0.2,
    n_jobs=-1,
    random_state=2
)

eval_set = [(X_train_encoded, y_train), 
            (X_val_encoded, y_val)]

model.fit(X_train_encoded, y_train, 
          eval_set=eval_set,
          eval_metric='mae', # #(wrong cases)/#(all cases)
          early_stopping_rounds=50
         ) # 50 rounds 동안 스코어의 개선이 없으면 멈춤
```

```python
#개선된 모델 성능 확인
y_train_pred_xg=model.predict(X_train_encoded)
y_val_pred_xg=model.predict(X_val_encoded)


print('훈련 데이터 MAE :', mean_absolute_error(y_train, y_train_pred_xg))
print('훈련 데이터 R2 score :', r2_score(y_train, y_train_pred_xg))

print('검증 데이터 MAE :', mean_absolute_error(y_val, y_val_pred_xg))
print('검증 데이터 R2 score :', r2_score(y_val, y_val_pred_xg))
```
